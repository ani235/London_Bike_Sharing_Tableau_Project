```python
import pandas as pd
```

```python
df = pd.read_csv('/content/london_merged.csv')
```

```python
#counting the unique values in the weather_code column
df['weather_code'].value_counts()
```

```python
#counting the unique values in the season column
df['season'].value_counts()
```

```python
#specifying new column names

cols_dict = {
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'percent_humidity',
    'windspeed':'windspeed_kph',
    'weather_code': 'weather'
}
```

```python
#Renaming tghe columns to the specified column names

df.rename(cols_dict, axis = 1, inplace = True)
```

```python
df
```

```python
# changing humidity values to between 0 and 1 by dividingby 100:

df['percent_humidity'] = df['percent_humidity']/100
```

```python
df
```

```python
# creating season dictionary so we can map the integers to actual written values

season_dict = {
    '0.0' : 'spring',
    '1.0' : 'summer',
    '2.0' : 'autumn',
    '3.0' : 'winter'
}
```

```python
# creating weather dictionary so we can map the integers to actual written values

weather_dict = {
    '1.0': 'Clear',
    '2.0': 'Scattered Clouds',
    '3.0': 'Broken Clouds',
    '4.0': 'Cloudy',
    '7.0': 'Rain',
    '10.0': 'Rain with Thunderstorms',
    '26.0': 'Snowfall'
}
```

```python
#changing the season column data type to string
df['season'] = df['season'].astype('str')
```

```python
# mapping the values of the season column to the actual values
df['season'] = df['season'].map(season_dict)
```

```python
#changing the weather column data type to string
df['weather'] = df['weather'].astype('str')
```

```python
#mapping the values of the weather column to the actual values
df['weather'] = df['weather'].map(weather_dict)
```
